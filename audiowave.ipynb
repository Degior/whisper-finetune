```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
from datasets import load_dataset, Audio
```

```python
dataset = load_dataset("mozilla-foundation/common_voice_11_0", "ru", split="test")
```

```python
dataset = dataset.cast_column("audio", Audio(sampling_rate=16000))
```

```python
samples = dataset.select(range(100))
```

```python
example_indices = [0, 5, 10]
```

```python
plt.figure(figsize=(18, 10))
```

```python
plt.figure(figsize=(18, 20))
for i, idx in enumerate(example_indices):
    example = samples[idx]
    audio = example["audio"]["array"]
    sampling_rate = example["audio"]["sampling_rate"]

    plt.subplot(3, len(example_indices), i + 1)
    librosa.display.waveshow(audio, sr=sampling_rate)
    plt.title(f"Waveform: Sample {idx}")
    plt.xlabel("Time (s)")
    plt.ylabel("Amplitude")
plt.tight_layout()
```

```python
plt.figure(figsize=(18, 20))
for i, idx in enumerate(example_indices):
    example = samples[idx]
    audio = example["audio"]["array"]
    sampling_rate = example["audio"]["sampling_rate"]

    plt.subplot(3, len(example_indices), i + 1)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(audio)), ref=np.max)
    librosa.display.specshow(D, sr=sampling_rate, x_axis='time', y_axis='log')
    plt.colorbar(format="%+2.0f dB")
    plt.title(f"Spectrogram: Sample {idx}")
    plt.xlabel("Time (s)")
    plt.ylabel("Frequency (Hz)")
plt.tight_layout()
```

```python
plt.figure(figsize=(18, 20))
for i, idx in enumerate(example_indices):
    example = samples[idx]
    audio = example["audio"]["array"]
    sampling_rate = example["audio"]["sampling_rate"]

    plt.subplot(3, len(example_indices), i + 1)
    S = librosa.feature.melspectrogram(y=audio, sr=sampling_rate, n_mels=128, fmax=8000)
    S_dB = librosa.power_to_db(S, ref=np.max)
    librosa.display.specshow(S_dB, sr=sampling_rate, x_axis='time', y_axis='mel', fmax=8000)
    plt.colorbar(format="%+2.0f dB")
    plt.title(f"Mel-Spectrogram: Sample {idx}")
    plt.xlabel("Time (s)")
    plt.ylabel("Mel frequency")
plt.tight_layout()
```
